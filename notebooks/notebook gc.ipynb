```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.lines import Line2D
from pathlib import Path
```

```python
PAR_DIR = Path('__file__').parent.parent
txtpath = f'{PAR_DIR}\\datasets\\GCUpalermo_process.txt'
txtpath1 = f'{PAR_DIR}\\datasets\\GCUNjujuy_process.txt'
```

```python
df_p = pd.read_csv(txtpath)
df_p.head()
```

```python
df_p.columns[1:]
```

```python
df_j = pd.read_csv(txtpath1)
df_j.head()
```

```python
df_j.columns[1:]
```

```python
df_p['career'].value_counts().plot.barh(color = 'g')
plt.gca().set(title='Distribución de carreras UP')
```

```python
df_j['career'].value_counts().plot.barh()
plt.gca().set(title='Distribución de carreras UNJ')
```

```python
plt.hist(df_p['age'], bins=100, color='g')
plt.gca().set(title='Distribución de edades UP', ylabel='Frequencia absoluta', xlabel='Edad')
```

```python

plt.hist(df_j[df_j.age != 59]['age'], bins=100)
plt.gca().set(title='Distribución de edades UNJ', ylabel='Frequencia absoluta', xlabel='Edad')
```

```python
fig, ax = plt.subplots()
bars = ax.bar(df_p['gender'].value_counts().index, height = df_p['gender'].value_counts().values, color = 'g')
plt.title('Distribución de géneros UP')

ax.bar_label(bars)
```

```python
fig, ax = plt.subplots()
bars = ax.bar(df_j['gender'].value_counts().index, height = df_j['gender'].value_counts().values)
plt.title('Distribución de géneros UNJ')
ax.bar_label(bars)
```

```python
localidades = []
frecuencia = []

for n in range (10,0,-1):
    num, loc = n, len(df_p.groupby('location').filter(lambda x: len(x) == n)['location'].value_counts().index)
    localidades.append(loc)
    frecuencia.append(num)
```

```python
fig, ax = plt.subplots()
bars = ax.bar(frecuencia, height = localidades, color='g')
plt.title('Frecuencia de localidades UP')
ax.bar_label(bars)
```

```python
localidades1 = []
frecuencia1 = []

for n in range (10,0,-1):
    num, loc = n, len(df_j.groupby('location').filter(lambda x: len(x) == n)['location'].value_counts().index)
    localidades1.append(loc)
    frecuencia1.append(num)
```

```python
fig, ax = plt.subplots()
bars1 = ax.bar(frecuencia1, height = localidades1)
plt.title('Frecuencia de localidades UNJ')

ax.bar_label(bars1)
```

```python
df_p_carreras_unique = list(df_p['career'].unique())
df_p_carreras_unique.sort(key=lambda x: x)
df_j_carreras_unique = list(df_j['career'].unique())
df_j_carreras_unique.sort(key=lambda x: x)
df_both_carreras_unique = pd.DataFrame([z for z in zip(df_p_carreras_unique, df_j_carreras_unique)], columns=['Palermo', 'Jujuy'])
```

```python
df_both_carreras_unique
```

```python
df_both = pd.concat([df_p, df_j])
df_both_carrera = df_both[['career', 'university', 'last_name']].groupby(['career', 'university']).count().sort_values('last_name', ascending=False)
plt.figure(figsize=(13,18))
plt.title('Inscripciones UP vs UNJ - Por carrera')
ax = sns.countplot(data=df_both, y='career', hue='university', order=df_p_carreras_unique)
ax.bar_label(ax.containers[0], label_type='edge')
ax.bar_label(ax.containers[1], label_type='edge')
plt.show()
```

```python
df_both = pd.concat([df_p, df_j])
df_both_carrera = df_both[['career', 'university', 'last_name']].groupby(['career', 'university']).count().sort_values('last_name', ascending=False)
plt.figure(figsize=(10,8))
plt.title('Inscripciones UP vs UNJ - Por año')
df_both['inscription_date'] = pd.to_datetime(df_both['inscription_date'], format='%Y-%m-%d')
df_both['inscription_year'] = df_both['inscription_date'].map(lambda x: x.year)
ax = sns.countplot(data=df_both, x='inscription_year', hue='university')
ax.bar_label(ax.containers[0], label_type='edge')
ax.bar_label(ax.containers[1], label_type='edge')
plt.show()
```

```python
plt.figure(figsize=(12,9))
plt.title('Inscripciones UP vs UNJ - Por género')
ax = sns.countplot(data=df_both, x='gender', hue='university')
ax.bar_label(ax.containers[0], label_type='edge')
ax.bar_label(ax.containers[1], label_type='edge')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

colors = {'male':'tab:blue', 'female':'tab:orange'}

scatter = ax.scatter(df_p['age'], df_p['career'], c=df_p['gender'].map(colors), label = ['male', 'female'])
handles = [Line2D([0], [0], marker='o', color='w', markerfacecolor=v, label=k, markersize=8) for k, v in colors.items()]
ax.legend(title='gender', handles=handles, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xlabel("Edad", fontsize = 15)
plt.ylabel("Carreras", fontsize = 15)
plt.title('Asociación entre carrera, edad y genero UP', fontsize = 18)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

colors = {'male':'tab:blue', 'female':'tab:orange'}

scatter = ax.scatter(df_p['age'], df_p['location'], c=df_p['gender'].map(colors), label = ['male', 'female'])
handles = [Line2D([0], [0], marker='o', color='w', markerfacecolor=v, label=k, markersize=8) for k, v in colors.items()]
ax.legend(title='gender', handles=handles, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xlabel("Edad", fontsize = 15)
plt.ylabel("Localidades", fontsize = 15)
plt.title('Asociación entre localidad, edad y genero UP', fontsize = 18)
plt.tick_params(axis='y', width =1, which='major', labelsize=2)
plt.show()
```
